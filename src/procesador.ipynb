## ESTRUCTURA Y CARGA DE DATOS 

```python
import pandas as pd
from typing import List
```

```python
class ProcesadorEconomico:

    def __init__(self):
        self.df = pd.DataFrame()

    def cargar_datos(self, ruta_json: str) -> pd.DataFrame:
        """
        Convierte un archivo JSON en DataFrame.
        """
        df = pd.read_json(ruta_json)
        return df
        
    def unir_datasets(self, lista_dataframes: List[pd.DataFrame]) -> pd.DataFrame:
        
        self.df = pd.concat(lista_dataframes, ignore_index=True)
        return self.df
```

## LIMPIEZA MÁS REGEX

```python
def limpiar_y_transformar(self) -> pd.DataFrame:

    df = self.df.copy()

    df["anio"] = df["fecha_original"].astype(str).str.extract(r"(\d{4})")
    df["anio"] = df["anio"].astype(int)

    df["valor"] = pd.to_numeric(df["valor"], errors="coerce")

    df = df.dropna(subset=["valor"])

    df = df.sort_values(by=["pais", "indicador", "anio"])

    self.df = df
    return df
```
